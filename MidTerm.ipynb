# Pre-Processing Pipeline

```python
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
pipeline = [str.lower, tokenize]
def prepare(reviewtext, pipeline):
    tokens = reviewtext 
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
#Run the pipeline
df['tokens'] = df['reviewText'].apply(prepare, pipeline=pipeline)
df
```

```python
# importing a function Counter
from collections import Counter

counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq']) # word frequency in itâ€™s owndataframe for analysis
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

# Sentimental Analysis

```python
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) #Remove @mentions replace with blank
    text = re.sub(r'#', '', text) #Remove the '#' symbol, replace with blank
    text = re.sub(r'RT[\s]+', '', text) #Removing RT, replace with blank
    text = re.sub(r'https?:\/\/\S+', '', text) #Remove the hyperlinks
    text = re.sub(r':', '', text) # Remove :
    
    return text

#Cleaning the text
df["reviewText"]= df["reviewText"].apply(cleanTxt)

#Show the clean text
df.head()
```

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

#Cleaning the text
df["reviewText"]= df["reviewText"].apply(remove_emoji)

#Show the clean text
df.head()
```

# Bing Liu Score

```python
from sklearn import preprocessing
import nltk
nltk.download('opinion_lexicon')
from nltk.corpus import opinion_lexicon
from nltk.tokenize import word_tokenize
print('Total number of words in opinion lexicon', 
      len(opinion_lexicon.words()))
print('Examples of positive words in opinion lexicon',
      opinion_lexicon.positive()[:10])
print('Examples of negative words in opinion lexicon',
      opinion_lexicon.negative()[:10])
```

```python
# Let's create a dictionary which we can use for scoring our review content

# Downloading punkt from NLTK library
nltk.download('punkt')

# Renaming the column 'reviewcontent' to 'CONTENT' in the DataFrame
df.rename(columns={"review": "reviewText"}, inplace=True)

# Assigning positive and negative scores
pos_score = 1
neg_score = -1

# Initializing an empty dictionary
word_dict = {}
 
# Adding the positive words to the dictionary
for word in opinion_lexicon.positive():
        word_dict[word] = pos_score
      
# Adding the negative words to the dictionary
for word in opinion_lexicon.negative():
        word_dict[word] = neg_score
```

```python
#Creating a fuunction content
def bing_liu_score(reviewText):
    #Initializing the sentiment score
    sentiment_score = 0
    #Tokenizing the input conetnt into words and convert them to lowercase
    bag_of_words = word_tokenize(reviewText.lower())
     # creating loop to check each word in the bag of words
    for word in bag_of_words:
        #Checking if the word exists in the sentiment dictionary
        if word in word_dict:
            # If the word exist, adding its sentiment score to the sentiment score
            sentiment_score += word_dict[word]
    return sentiment_score  #Returning the sentiment score for the content
```
