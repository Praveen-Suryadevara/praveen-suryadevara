# Tokenization

```python
#Tokenise a given text based on a regular expression pattern
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

# Testing 

```python
#Testing the tokenise with a testtext
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

# Removing 

```python
#How to remove stop words
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

# Funtion for Removing

```python
#Function to remove stop words
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

# Pipeline

```python
#Making a Pipeline
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text 
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Execution

```python
#Run the pipeline
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

# Counting

```python
#Count the tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Counters

```python
from collections import Counter 
tokens = tokenize("She likes my cats and my cats like my sofa.") # Using the Counter
testcount=Counter(tokens)
print(testcount)
```

# Adding Tokens

```python
#Adding more tokens
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

# Counting Tokens

```python
#Count the tokens
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

# Analysis

```python
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq']) # word frequency in it’s owndataframe for analysis
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

# Sort & Display

```python
#Sorting and displaying 
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

# Bar-Chart

```python
#Creating a bar chart
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
